Create a basic graph
The networkx package is used to create a basic graph

*Import the pacakge*   
*Use inline plots*  

```python
import networkx as nx
%matplotlib inline
```

Create basic graph by using the list of nodes and the links between nodes. Below example creates nodes first and then provides edges. Alternatively just edges could be provided and the networkx will add nodes using the nodes on the edges list.

```python
G=nx.Graph()
people=['Andre','Beverly','Carol','Diane','Ed','Fernando','Garth','Heather','Ike','Jane']
G.add_nodes_from(people)

links=[('Andre','Beverly'),('Andre','Carol'),('Andre','Diane'),('Andre','Fernando'),
       ('Beverly','Diane'),('Beverly','Ed'),('Beverly','Garth'),
       ('Carol','Diane'),('Carol','Fernando'),
       ('Diane','Ed'),('Diane','Fernando'),('Diane','Garth'),
       ('Ed','Garth'),
       ('Fernando','Garth'),('Fernando','Heather'),
       ('Garth','Heather'),
       ('Heather','Ike'),
       ('Ike','Jane')]
G.add_edges_from(links)

G.nodes()
```

```python
G.edges()
```

#### Graph Diagram
Let's draw the graph

```python
nx.draw(G,with_labels=True,font_size=16)               
```

#### Degree 
Let's try to visualize the number of connection each person has by changing the size of the node based on number of connections

```python
deg=nx.degree(G)
deg
node_size = [d * 200 for d in deg.values()]
nx.draw(G,with_labels=True,font_size=16,node_size=node_size) 
```
